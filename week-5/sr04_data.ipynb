```python
import random
import datetime
import json

def generate_sr04_data(num_entries=10):
    data = []
    current_time = datetime.datetime.now()

    for _ in range(num_entries):
        entry = {
            "timestamp": current_time.isoformat(),
            "measurements": {
                "distance": round(random.uniform(2, 400), 1)  # SR04 typically measures 2-400 cm
            }
        }
        data.append(entry)
        current_time += datetime.timedelta(seconds=30)  # Readings every 30 seconds

    sensor_data = {
        "sensor_name": "SR04",
        "sensor_type": "ultrasonic",
        "data": data
    }

    return sensor_data

# Generate and print the data
random_data = generate_sr04_data()
print(json.dumps(random_data, indent=2))
```
